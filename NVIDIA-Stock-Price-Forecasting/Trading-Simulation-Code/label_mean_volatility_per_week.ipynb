```python
import pandas as pd
```

```python
df_reference = pd.read_csv("/content/drive/MyDrive/Data Science Eugene Pinsky/Stock Market/NVDA_with_labels.csv")
df_reference.head()
```

```python
df_summary = pd.read_csv("/Volumes/GoogleDrive/My Drive/Data Science Eugene Pinsky/Stock Market/NVDA.csv")
df_summary.shape
```

```python
#appending label column to summary data using reference data

#combining duplicate week number columns into a single unique week number column.
g = df_reference['Week_Number'].ne(df_reference['Week_Number'].shift()).cumsum() 
df_temp = df_reference.groupby(g).first()
```

```python
df_temp
```

```python
df_temp = df_temp.reset_index(drop=True)
df_temp.head()
```

```python
label_column = df_temp["Label"]

df_summary = df_summary.join(label_column)
```

```python
df_summary.to_csv("NVDA_mean_volatility_labels.csv",index=False)
```
